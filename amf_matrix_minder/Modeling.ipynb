```python
def combine(pod):
    value = pod.select(explode(pod.values).alias('values'),'metric_id')
    time = pod.select(explode(pod.date_col).alias('date_col'))
    df_merged = value.join(time)
    df_merged = df_merged.withColumn('date_col', to_timestamp('date_col'))
    value = value.toPandas()
    time = time.toPandas()
    values= time.iloc[:, 0].values
    value['date_col'] = values
    spark = SparkSession.builder.getOrCreate()
    combined_df = spark.createDataFrame(value)
    return combined_df
```

```python
def process_df(df, feature):
    df_date = df.withColumn("date_col",  expr("transform(timestamps, x -> from_unixtime(x / 1000))"))
    pod = df_date.filter(col("metric_pod").startswith('open5gs-amf')).select("metric_pod",'date_col','metric___name__','values','metric_id').filter((col('metric___name__')==feature))
    df_pod = combine(pod)


    df_pod = df_pod.filter(col("metric_id") == "/kubepods.slice/kubepods-besteffort.slice/kubepods-besteffort-pod359da734_d215_4ea7_9e05_5e7bdd7d4b07.slice/cri-containerd-984ec0cb71c7442cdab41f4be47e75803ee54d4b3a9a33e1391a3f678643cda0.scope").select('values','date_col').withColumnRenamed("values", feature).withColumnRenamed("date_col", "timestamp")

    return datetime(df_pod.toPandas())
```

```python

from scipy.spatial import cKDTree

# Left join df_warning on df2_p based on the 'timestamp' column
df_join = pd.merge(df2_p, df_warning, on='timestamp', how='left')

# Get the timestamps in df_warning that are not in df2_p
missing_timestamps = set(df_warning['timestamp']) - set(df2_p['timestamp'])

def group(df,df_join):
    if len(missing_timestamps) > 0:
        df2_p_timestamps = pd.to_datetime(df['timestamp'])
        df_warning_timestamps = pd.to_datetime(df_warning['timestamp'])
        tree = cKDTree(df2_p_timestamps.values.reshape(-1, 1))

        nearest_indices = tree.query(df_warning_timestamps.values.reshape(-1, 1))[1]
        missing_values_dict = {column: df_warning[column].values for column in df_warning.columns[1:]}
        df_merged_missing = df.iloc[nearest_indices].copy()
        for column, values in missing_values_dict.items():
            df_merged_missing[column] = values

        df_join = pd.concat([df_join, df_merged_missing], ignore_index=True)
    # Sort the merged dataframe by the 'timestamp' column
    df_join.sort_values('timestamp', inplace=True)

    # Reset the index of the merged dataframe
    df_join.reset_index(drop=True, inplace=True)
    df_join.fillna(0, inplace=True)


    # Group by 'timestamp' and sum the values in the other columns
    df_merged_grouped = df_join.groupby('timestamp').sum()

    # Reset the index to make the 'timestamp' column a regular column
    df_merged_grouped.reset_index(inplace=True)

    # Sort the merged dataframe by the 'timestamp' column
    df_merged_grouped.sort_values('timestamp', inplace=True)

    # Reset the index of the merged dataframe
    df_merged_grouped.reset_index(drop=True, inplace=True)
    return df_merged_grouped

df_merged_grouped = group(df2_p,df_join)
```

```python
def merge(df,df_join, df_merged_grouped):
    if len(missing_timestamps) > 0:
        df2_p_timestamps = pd.to_datetime(df['timestamp'])
        df_warning_timestamps = pd.to_datetime(df_warning['timestamp'])
    
        tree = cKDTree(df2_p_timestamps.values.reshape(-1, 1))

        nearest_indices = tree.query(df_warning_timestamps.values.reshape(-1, 1))[1]

        missing_values_dict = {column: df_warning[column].values for column in df_warning.columns[1:]}

        df_merged_missing = df.iloc[nearest_indices].copy()
        for column, values in missing_values_dict.items():
            df_merged_missing[column] = values

        df_merged = pd.concat([df_join, df_merged_missing], ignore_index=True)

    df_merged.sort_values('timestamp', inplace=True)

    df_merged.reset_index(drop=True, inplace=True)

    columns_to_sum = ['Registration reject', 'Cannot find SUCI', 'Authentication reject',
                  'Registration request', 'GUTI has already been allocated',
                  '5GMM STATUS', 'OLD UE Context Release']

    df_merged_grouped = df_merged.groupby('timestamp')[columns_to_sum].sum().reset_index()
    df_merged = pd.merge(df, df_merged_grouped, on='timestamp', how='left')
    df_merged[['Registration reject', 'Cannot find SUCI', 'Authentication reject', 'Registration request', 'GUTI has already been allocated', '5GMM STATUS', 'OLD UE Context Release']] = df_merged_grouped[['Registration reject', 'Cannot find SUCI', 'Authentication reject', 'Registration request', 'GUTI has already been allocated', '5GMM STATUS', 'OLD UE Context Release']].applymap(lambda x: 1 if x > 0 else 0)
    return df_merged



df_merged = merge(df2_p,df_join,df_merged_grouped)
```

```python
from pyspark.sql import SparkSession
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matrixprofile as mp

# start a Spark Session
spark = SparkSession.builder.getOrCreate()

dataset = df_merged2['container_memory_cache'].values
timestamps = df_merged2['timestamp'].values
window_sizes = [8, 20, 40, 60]  # These are in half-minutes
distance_thresholds = [3.5, 6,1.9, 1]  

# Create a figure with multiple subplots
fig, axs = plt.subplots(len(window_sizes), 1, sharex=True, figsize=(20,10))

for i, (window_size, threshold) in enumerate(zip(window_sizes, distance_thresholds)):
    profile = mp.compute(dataset, window_size)
    motif = np.where(profile['mp'] < threshold)[0]
    axs[i].plot(profile['mp'], label='Matrix profile', color='blue')
    axs[i].plot(motif, profile['mp'][motif], 'o', label='Anomalies', color='red')
    axs[i].legend()
    axs[i].set_title(f'Window size: {window_size}, Distance threshold: {threshold}')
    axs[i].set_ylabel('Matrix Profile')
    axs[i].set_ylim(0, max(profile['mp']) + 1)
    motif_timestamps = timestamps[motif]
    
    # Calculate end timestamps for each anomaly
    anomaly_intervals = [(start, start + pd.Timedelta(minutes=window_size*0.5)) for start in pd.to_datetime(motif_timestamps)]
    anomaly_intervals = [(start.to_pydatetime(), end.to_pydatetime()) for start, end in anomaly_intervals]



# Set common labels
fig.text(0.5, 0.04, 'Time', ha='center')
plt.tight_layout()
plt.show()
```
